```python
import numpy as np

A = np.array([[1,2,3], [1,2,3], [1,2,3]])
b = np.array([4,5,6])

print(A@b)

np.einsum('ij,j->i', A, b)
```

```python
def cross_product(a,b):
    return np.array([a[1]*b[2] - a[2]*b[1],
                     a[2]*b[0] - a[0]*b[2],
                     a[1]*b[0] - a[0]*b[1]])

cross_product(np.array([1,2,3]), np.array([4,5,6]))
```

```python
%timeit cross_product(np.array([1,2,3]), np.array([4,5,6]))
```

```python
%timeit np.cross(np.array([1,2,3]), np.array([4,5,6]))
```
