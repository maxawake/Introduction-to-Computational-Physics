```python
import sys

import matplotlib.pyplot as plt
import maxpy
import numpy as np

A = 2
lamb = 1


def potential(x, lamb):
    return 0.5 * x**2 + lamb / 4 * x**4


x = np.linspace(-A, A, 100)
V = potential(x, lamb)

plt.plot(x, V)
plt.xlabel("x")
plt.ylabel("V(x)")
plt.title("Potential Energy Function")
plt.grid()
```

```python
def integrand(x, A, lamb):
    return 1 / np.sqrt(2 * (potential(A, lamb) - potential(x, lamb)))


plt.plot(x, integrand(x, A, lamb))
```

```python
N = 1000000
xlin = np.linspace(0, A, N)
h = xlin[1] - xlin[0]


def open_midpoint(h, xlin, A, lamb):
    return h * np.sum(integrand(xlin[:-1] + h / 2, A, lamb))


print("Open Midpoint:", 4 * open_midpoint(h, xlin, A, lamb))
```

```python
import scipy.integrate as integrate


def integrand_quad(x, A, lamb):
    return 1 / np.sqrt(2 * (potential(A, lamb) - potential(x, lamb)))


result, error = integrate.quad(integrand_quad, 0, A, args=(A, lamb))
print("Quad Result:", result)
```
